```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('DiwaliSales.csv' , encoding = 'unicode_escape')
# to avoid the encoding error in file
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1 , inplace = True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
ax = sns.countplot(x = 'Gender' , data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'] , as_index = False ) ['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.barplot(x = 'Gender' , y = 'Amount' , data = sales_gen)
```

```python
# from above graph we can see women have purchesed more items than men .
```

```python
ax = sns.countplot(data=df , x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
# from the above graph we can see that most of the buyers are lies b/t 26 - 35
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(11)
sns.set(rc={'figure.figsize': (20, 7)})
sns.barplot(data=sales_state, x='State', y='Orders')
# total no. of oders from top 10 states
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (18, 7)})
sns.barplot(data=sales_state, x='State', y='Amount')
# total no. of Amount spended from top 10 states
```

```python
ax = sns.countplot(data = df , x = 'Marital_Status')

sns.set(rc = {'figure.figsize':(7,10 )})
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (10, 10)})
sns.barplot(data=sales_state, x='Marital_Status', y='Amount' , hue = 'Gender')
```
